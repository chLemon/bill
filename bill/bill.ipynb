```python
from TradeRecord import TradeRecord, read_trade_records
from pathlib import Path
import read_ali
import read_wechat
```

```python
from enum import Enum
class Bill(Enum):
    WECHAT = (read_wechat.read_wx_file, "../data/微信", read_wechat.BILL_TYPE)
    ALIPAY = (read_ali.read_ali_file, "../data/支付宝", read_ali.BILL_TYPE)

    def __init__(self, read_func, data_path, bill_name):
        self.read_func = read_func
        self.data_path = data_path
        self.bill_name = bill_name
```

```python
all_trade_records = []
times_map = {}
for bill in Bill:
    trade_records, times = read_trade_records(bill.read_func, Path(bill.data_path))
    all_trade_records.extend(trade_records)
    times_map[bill.bill_name] = times
```
