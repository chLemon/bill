```python
import csv
from domain.GroupData import GroupData

def write_excel(file_path, group_data):
    with open(file_path, 'w', newline='', encoding='utf-8-sig') as f:
        writer = csv.writer(f)
        for output in group_data.output_list:
            writer.writerow(output)
```

```python
for year, month_to_group_data in year_to_month_to_group_data.items():
    year_dir = Path(str(year) + "年账单")
    year_dir.mkdir(parents=True, exist_ok=True)
    for month, group_data in month_to_group_data.items():
        file_path = year_dir / (str(month) + "月账单.csv")
        write_excel(file_path, group_data)
```
