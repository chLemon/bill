```python
from datetime import datetime

def merge_time_ranges(time_list):
    # 将字符串转换为datetime对象，并排序
    intervals = sorted([(datetime.strptime(start, "%Y-%m-%d %H:%M:%S"),
                        datetime.strptime(end, "%Y-%m-%d %H:%M:%S")) for start, end in time_list])
    merged = []
    for interval in intervals:
        if not merged or merged[-1][1] < interval[0]:
            merged.append(list(interval))
        else:
            merged[-1][1] = max(merged[-1][1], interval[1])
    # 转回字符串格式
    return [(start.strftime("%Y-%m-%d %H:%M:%S"), end.strftime("%Y-%m-%d %H:%M:%S")) for start, end in merged]

# 示例
time_list = [
    ("2025-04-07 00:00:00", "2025-04-09 00:00:00"),
    ("2025-04-08 12:00:00", "2025-04-10 00:00:00"),
    ("2025-04-11 00:00:00", "2025-04-12 00:00:00")
]
merged_time_list = merge_time_ranges(time_list)
merged_time_list
```
