```python
dir_path = "../../data/支付宝"
```

```python
from pathlib import Path

import sys

# setting path
sys.path.append('../')

from TradeRecord import TradeRecord
```

```python
def read_ali_file(file):
    with open(file, 'r', encoding='gb2312') as f:
        trade_record_str_list = f.readlines()[25:]
    return [TradeRecord.from_ali_str(trade_record_str) for trade_record_str in trade_record_str_list]


trade_record_list = []

files = [file for file in Path(dir_path).iterdir() if file.is_file()]

for file in files:
    trade_record_list.extend(read_ali_file(file))
```
