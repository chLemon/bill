```python
dir_path = "../../../data/微信"
```

```python
from pathlib import Path

import sys

# setting path
sys.path.append('../../')

from domain.TradeRecord import TradeRecord

files = [file for file in Path(dir_path).iterdir() if file.is_file()]

def read_wx_file(file):
    with open(file, 'r', encoding='utf-8') as f:
        trade_record_str_list = f.readlines()[17:]
    return [TradeRecord.from_wx_str(trade_record_str, "微信") for trade_record_str in trade_record_str_list]

trade_record_list = []

for file in files:
    trade_record_list.extend(read_wx_file(file))

```
