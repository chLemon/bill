```python
dir_path = "../../../data/浦发银行/每月的对账单"
files = [file for file in Path(dir_path).iterdir() if file.is_file()]
pdf_file = files[0]
```

```python
from pypdf import PdfReader

reader = PdfReader(pdf_file)
text = ""
for page in reader.pages:
    text += page.extract_text()
```
